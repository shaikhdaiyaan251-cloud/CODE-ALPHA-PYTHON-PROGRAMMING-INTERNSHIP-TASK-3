## **4.1. Setting Up the Colab Environment**
 4.1.1 Import required modules (all from standard library)

 4.1.2 Define coloured output helper (ANSI escape codes)

 4.1.3 Typing simulation effect (optional, makes bot feel more natural)

 4.1.4 Display welcome message and instructions


````python
# ==================================================
# 4.1. Setting Up the Colab Environment – Basic Chatbot
# ==================================================

# ----------------------------
# 4.1.1 Import required modules (all from standard library)
# ----------------------------
import time
import random
import sys

# ----------------------------
# 4.1.2 Define coloured output helper (ANSI escape codes)
# ----------------------------
def print_colored(text, color='white', bold=False):
    """Print text with ANSI colour and optional bold (works in Colab)."""
    colors = {
        'red': '\033[91m', 'green': '\033[92m', 'yellow': '\033[93m',
        'blue': '\033[94m', 'purple': '\033[95m', 'cyan': '\033[96m',
        'white': '\033[97m', 'reset': '\033[0m'
    }
    bold_code = '\033[1m' if bold else ''
    reset_code = colors['reset']
    color_code = colors.get(color, colors['white'])
    print(f"{bold_code}{color_code}{text}{reset_code}")

# ----------------------------
# 4.1.3 Typing simulation effect (optional, makes bot feel more natural)
# ----------------------------
def type_like_bot(text, delay=0.03):
    """Print text character by character to simulate typing."""
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()  # newline after typing

# ----------------------------
# 4.1.4 Display welcome message and instructions
# ----------------------------
print_colored("=" * 50, 'cyan', bold=True)
print_colored("🤖 WELCOME TO THE BASIC CHATBOT! 🤖", 'green', bold=True)
print_colored("=" * 50, 'cyan', bold=True)

print("\n📜 INSTRUCTIONS:")
print("   • Type your messages and press Enter.")
print("   • The bot will respond based on keywords.")
print("   • Try saying: hello, how are you, your name, help, bye")
print("   • Type 'bye' or 'exit' to end the conversation.\n")

print_colored("✅ Environment ready! You can now run the chatbot loop.", 'green')
print("   Proceed to the next cell to start chatting.\n")
````

## **4.2. Defining the Rule‑Based Response Logic**
 4.2.1 Response dictionary (keyword → reply)

 4.2.2 Default response when no keyword matches

 4.2.3 Function to get bot response based on user input

 4.2.4 (Optional) Test the response logic


````python
# ==================================================
# 4.2. Defining the Rule‑Based Response Logic
# ==================================================

# ----------------------------
# 4.2.1 Response dictionary (keyword → reply)
# ----------------------------
responses = {
    # Greetings
    "hello": "Hi there! 😊 How can I help you today?",
    "hi": "Hello! 👋 What's on your mind?",
    "hey": "Hey! 😄 Ready to chat?",

    # How are you variations
    "how are you": "I'm just a bunch of code, but I'm doing great! Thanks for asking. 🤖",
    "how are u": "I'm functioning perfectly! How about you? 😊",
    "how do you do": "I do well, thank you! And you?",

    # User name / identity
    "your name": "I'm ChatBot, your virtual assistant! 🤖",
    "who are you": "I'm a simple rule‑based chatbot created for CodeAlpha. Nice to meet you!",

    # Bot's well‑being (joke)
    "are you real": "As real as Python code can be! 😉",
    "what can you do": "I can chat about simple topics. Try saying 'hello', 'how are you', or 'bye'!",

    # Help
    "help": "I understand greetings, 'how are you', 'your name', 'bye', and a few more. Just type naturally!",

    # Farewells (will be used to exit loop)
    "bye": "Goodbye! Have a wonderful day! 👋",
    "exit": "Exiting chat. Take care! 👋",
    "quit": "Quitting. See you next time! 👋",

    # Small talk
    "thanks": "You're very welcome! 😊",
    "thank you": "My pleasure! 🤗"
}

# ----------------------------
# 4.2.2 Default response when no keyword matches
# ----------------------------
DEFAULT_RESPONSE = "I'm not sure how to respond to that. 🤔 Try saying 'hello', 'how are you', or 'help'."

# ----------------------------
# 4.2.3 Function to get bot response based on user input
# ----------------------------
def get_bot_response(user_input):
    """
    Match user input against response keywords and return the corresponding reply.

    Parameters:
    user_input (str): The user's message (already lowercased and stripped).

    Returns:
    str: The bot's reply.
    """
    # Check each keyword to see if it appears in the user input
    for keyword, reply in responses.items():
        if keyword in user_input:
            return reply
    # If no keyword matches, return default
    return DEFAULT_RESPONSE

# ----------------------------
# 4.2.4 (Optional) Test the response logic
# ----------------------------
if __name__ == "__main__":
    test_phrases = [
        "hello",
        "Hello there!",
        "how are you",
        "how are u doing?",
        "your name?",
        "what can you do?",
        "bye bye",
        "random stuff"
    ]
    print("Testing response logic:\n")
    for phrase in test_phrases:
        reply = get_bot_response(phrase.lower())
        print(f"User: {phrase}")
        print(f"Bot: {reply}\n")
````

## **4.3. Implementing the Chat Loop**
 4.3.1 Helper function to clean user input

 4.3.2 Main chat loop

 4.3.3 Run the chatbot (if this script is executed directly)


````python
# ==================================================
# 4.3. Implementing the Chat Loop
# ==================================================

# ----------------------------
# 4.3.1 Helper function to clean user input
# ----------------------------
def clean_input(text):
    """Convert to lowercase, strip extra spaces, remove punctuation."""
    # Remove common punctuation that might interfere with matching
    punctuation = ".,!?;:'\""
    for p in punctuation:
        text = text.replace(p, '')
    return text.lower().strip()

# ----------------------------
# 4.3.2 Main chat loop
# ----------------------------
def chat():
    """Run the chatbot conversation loop."""
    print_colored("\n🤖 CHATBOT ACTIVATED. Start chatting! (type 'bye' to exit)\n", 'cyan')

    while True:
        # Get user input
        user_input = input("You: ").strip()

        # Exit if empty input (just re-prompt)
        if not user_input:
            print_colored("Bot: Please say something! 😊", 'yellow')
            continue

        # Clean the input for matching
        cleaned = clean_input(user_input)

        # Check for exit commands
        exit_words = ['bye', 'exit', 'quit']
        if cleaned in exit_words or any(word in cleaned for word in exit_words):
            # Get farewell response from dictionary
            farewell = get_bot_response(cleaned) if get_bot_response(cleaned) != DEFAULT_RESPONSE else "Goodbye! 👋"
            type_like_bot(f"Bot: {farewell}", delay=0.03)
            break

        # Get bot response (using function from Subtask 2)
        reply = get_bot_response(cleaned)

        # Display bot response with typing effect
        type_like_bot(f"Bot: {reply}", delay=0.03)

# ----------------------------
# 4.3.3 Run the chatbot (if this script is executed directly)
# ----------------------------
if __name__ == "__main__":
    chat()
````

## **4.4. Matching User Input to Responses**
 4.4.1 Intent mapping: multiple keywords per response

 4.4.2 Default response (no match)

 4.4.3 Function to find the best matching intent

 4.4.4 Function to check if input is a farewell (for loop exit)

 4.4.5 Test the matching

````python
# ==================================================
# 4.4. Matching User Input to Responses (Enhanced)
# ==================================================

# ----------------------------
# 4.4.1 Intent mapping: multiple keywords per response
# ----------------------------
# Instead of a flat dictionary, we define a list of intents.
# Each intent has:
#   - keywords: list of phrases that trigger this response
#   - response: the bot's reply (can also be a list for variety)
#   - priority: optional (higher = checked first)

intents = [
    {
        "keywords": ["hello", "hi", "hey", "greetings", "good morning", "good afternoon", "good evening"],
        "responses": [
            "Hi there! 😊 How can I help you today?",
            "Hello! 👋 What's on your mind?",
            "Hey! 😄 Ready to chat?"
        ],
        "priority": 1
    },
    {
        "keywords": ["how are you", "how are u", "how do you do", "how's it going", "how's everything"],
        "responses": [
            "I'm just a bunch of code, but I'm doing great! Thanks for asking. 🤖",
            "I'm functioning perfectly! How about you? 😊",
            "All systems operational! How can I assist you?"
        ],
        "priority": 2
    },
    {
        "keywords": ["your name", "who are you", "what are you", "introduce yourself"],
        "responses": [
            "I'm ChatBot, your virtual assistant! 🤖",
            "I'm a simple rule‑based chatbot created for CodeAlpha. Nice to meet you!"
        ],
        "priority": 1
    },
    {
        "keywords": ["are you real", "real", "alive"],
        "responses": [
            "As real as Python code can be! 😉",
            "I exist in the digital realm. Pretty real to me!"
        ],
        "priority": 1
    },
    {
        "keywords": ["what can you do", "capabilities", "features", "help me"],
        "responses": [
            "I can chat about simple topics. Try saying 'hello', 'how are you', or 'bye'!"
        ],
        "priority": 1
    },
    {
        "keywords": ["help", "commands", "what to say", "options"],
        "responses": [
            "I understand greetings, 'how are you', 'your name', 'bye', and a few more. Just type naturally!"
        ],
        "priority": 2
    },
    {
        "keywords": ["thanks", "thank you", "appreciate", "good bot"],
        "responses": [
            "You're very welcome! 😊",
            "My pleasure! 🤗",
            "Anytime! 😊"
        ],
        "priority": 1
    },
    {
        "keywords": ["bye", "goodbye", "exit", "quit", "see you", "farewell"],
        "responses": [
            "Goodbye! Have a wonderful day! 👋",
            "Exiting chat. Take care! 👋",
            "Quitting. See you next time! 👋"
        ],
        "priority": 3  # high priority to exit early
    }
]

# ----------------------------
# 4.4.2 Default response (no match)
# ----------------------------
DEFAULT_RESPONSE = "I'm not sure how to respond to that. 🤔 Try saying 'hello', 'how are you', or 'help'."

# ----------------------------
# 4.4.3 Function to find the best matching intent
# ----------------------------
def match_intent(user_input):
    """
    Find the best matching intent for the given user input.
    Returns the response string (randomised if multiple options) and the intent.
    """
    user_input_lower = user_input.lower()
    # Sort intents by priority (higher first)
    sorted_intents = sorted(intents, key=lambda x: x.get('priority', 0), reverse=True)

    for intent in sorted_intents:
        for keyword in intent["keywords"]:
            if keyword in user_input_lower:
                # Randomly select one response if multiple
                reply = random.choice(intent["responses"])
                return reply, intent
    return DEFAULT_RESPONSE, None

# ----------------------------
# 4.4.4 Function to check if input is a farewell (for loop exit)
# ----------------------------
def is_farewell(user_input):
    """Return True if the user input matches a goodbye intent."""
    reply, intent = match_intent(user_input)
    if intent and any(kw in user_input.lower() for kw in ["bye", "goodbye", "exit", "quit", "see you", "farewell"]):
        return True
    return False

# ----------------------------
# 4.4.5 Test the matching
# ----------------------------
if __name__ == "__main__":
    test_inputs = [
        "hello there!",
        "How are u doing?",
        "what's your name?",
        "can you help me?",
        "thanks a lot",
        "bye bye",
        "random gibberish"
    ]
    print("Testing intent matching:\n")
    for inp in test_inputs:
        reply, intent = match_intent(inp)
        print(f"User: {inp}")
        print(f"Bot: {reply}")
        print(f"Matched intent: {intent['keywords'][0] if intent else 'None'}\n")
````

## **4.5. Adding Polish & User Experience**
 4.5.1 Colour helper (with end parameter)

 4.5.2. Typing effect

 4.5.3. Conversation log

 4.5.4. Intents (rule‑based matching)

 4.5.5. Name extraction

 4.5.6. Bot speaking function (with logging)

 4.5.7. Main polished chat loop

 4.5.8. Run the chatbot


````python
# ==================================================
# 4.5. Adding Polish & User Experience
# ==================================================

import datetime
import random
import re
import time
import sys

# ----------------------------
# 4.5.1 Colour helper (with end parameter)
# ----------------------------
def print_colored(text, color='white', bold=False, end='\n'):
    """Print text with ANSI colour, optional bold, and custom end."""
    colors = {
        'red': '\033[91m', 'green': '\033[92m', 'yellow': '\033[93m',
        'blue': '\033[94m', 'purple': '\033[95m', 'cyan': '\033[96m',
        'white': '\033[97m', 'reset': '\033[0m'
    }
    bold_code = '\033[1m' if bold else ''
    reset_code = colors['reset']
    color_code = colors.get(color, colors['white'])
    print(f"{bold_code}{color_code}{text}{reset_code}", end=end)

# ----------------------------
# 4.5.2. Typing effect
# ----------------------------
def type_like_bot(text, delay=0.03):
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()

# ----------------------------
# 4.5.3. Conversation log
# ----------------------------
conversation_log = []

def log_message(sender, message):
    timestamp = datetime.datetime.now().strftime("%H:%M:%S")
    conversation_log.append(f"[{timestamp}] {sender}: {message}")

def save_conversation(filename=None):
    if not conversation_log:
        print_colored("No conversation to save.", 'yellow')
        return
    if filename is None:
        filename = f"chat_log_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
    with open(filename, 'w', encoding='utf-8') as f:
        f.write("\n".join(conversation_log))
    print_colored(f"💾 Conversation saved to {filename}", 'green')
    try:
        from google.colab import files
        files.download(filename)
    except:
        pass

# ----------------------------
# 4.5.4. Intents (rule‑based matching)
# ----------------------------
intents = [
    {
        "keywords": ["hello", "hi", "hey", "greetings", "good morning", "good afternoon", "good evening"],
        "responses": [
            "Hi there! 😊 How can I help you today?",
            "Hello! 👋 What's on your mind?",
            "Hey! 😄 Ready to chat?"
        ],
        "priority": 1
    },
    {
        "keywords": ["how are you", "how are u", "how do you do", "how's it going", "how's everything"],
        "responses": [
            "I'm just a bunch of code, but I'm doing great! Thanks for asking. 🤖",
            "I'm functioning perfectly! How about you? 😊",
            "All systems operational! How can I assist you?"
        ],
        "priority": 2
    },
    {
        "keywords": ["your name", "who are you", "what are you", "introduce yourself"],
        "responses": [
            "I'm ChatBot, your virtual assistant! 🤖",
            "I'm a simple rule‑based chatbot created for CodeAlpha. Nice to meet you!"
        ],
        "priority": 1
    },
    {
        "keywords": ["are you real", "real", "alive"],
        "responses": [
            "As real as Python code can be! 😉",
            "I exist in the digital realm. Pretty real to me!"
        ],
        "priority": 1
    },
    {
        "keywords": ["what can you do", "capabilities", "features", "help me"],
        "responses": [
            "I can chat about simple topics. Try saying 'hello', 'how are you', or 'bye'!"
        ],
        "priority": 1
    },
    {
        "keywords": ["help", "commands", "what to say", "options"],
        "responses": [
            "I understand greetings, 'how are you', 'your name', 'bye', and a few more. Just type naturally!"
        ],
        "priority": 2
    },
    {
        "keywords": ["thanks", "thank you", "appreciate", "good bot"],
        "responses": [
            "You're very welcome! 😊",
            "My pleasure! 🤗",
            "Anytime! 😊"
        ],
        "priority": 1
    },
    {
        "keywords": ["bye", "goodbye", "exit", "quit", "see you", "farewell"],
        "responses": [
            "Goodbye! Have a wonderful day! 👋",
            "Exiting chat. Take care! 👋",
            "Quitting. See you next time! 👋"
        ],
        "priority": 3
    }
]

DEFAULT_RESPONSE = "I'm not sure how to respond to that. 🤔 Try saying 'hello', 'how are you', or 'help'."

def match_intent(user_input):
    user_input_lower = user_input.lower()
    sorted_intents = sorted(intents, key=lambda x: x.get('priority', 0), reverse=True)
    for intent in sorted_intents:
        for keyword in intent["keywords"]:
            if keyword in user_input_lower:
                reply = random.choice(intent["responses"])
                return reply, intent
    return DEFAULT_RESPONSE, None

def is_farewell(user_input):
    reply, intent = match_intent(user_input)
    if intent and any(kw in user_input.lower() for kw in ["bye", "goodbye", "exit", "quit", "see you", "farewell"]):
        return True
    return False

# ----------------------------
# 4.5.5. Name extraction
# ----------------------------
user_name = None

def extract_name(user_input):
    patterns = [
        r"my name is (\w+)",
        r"i am (\w+)",
        r"call me (\w+)",
        r"i'm (\w+)"
    ]
    for pattern in patterns:
        match = re.search(pattern, user_input.lower())
        if match:
            return match.group(1).capitalize()
    return None

def get_personalised_greeting():
    if user_name:
        return f"Welcome back, {user_name}! 😊 How can I help you today?"
    else:
        return "Hi there! 😊 What's your name? (You can say 'My name is ...')"

# ----------------------------
# 4.5.6. Bot speaking function (with logging)
# ----------------------------
def bot_say(message, color='cyan', delay=0.03):
    print_colored("Bot: ", color, bold=True, end='')
    type_like_bot(message, delay)
    log_message("Bot", message)

# ----------------------------
# 4.5.7. Main polished chat loop
# ----------------------------
def chat_polished():
    global user_name
    print_colored("\n" + "="*50, 'cyan', bold=True)
    print_colored("🤖 WELCOME TO THE ADVANCED CHATBOT! 🤖", 'green', bold=True)
    print_colored("="*50, 'cyan', bold=True)
    print("\n💡 Tips:")
    print("   • Type 'save log' to save the conversation.")
    print("   • Type 'clear' to clear the screen (simulated).")
    print("   • Type 'help' for commands.")
    print("   • Type 'bye' to exit.\n")

    greeting = get_personalised_greeting()
    bot_say(greeting)

    while True:
        print_colored("You: ", 'green', end='')
        user_input = input().strip()

        if not user_input:
            bot_say("Please say something! 😊", 'yellow')
            continue

        log_message("You", user_input)

        cmd = user_input.lower()
        if cmd == 'save log':
            save_conversation()
            continue
        elif cmd == 'clear':
            print("\n" * 2)
            continue
        elif cmd == 'help':
            bot_say("I understand greetings, 'how are you', 'your name', 'bye'. Also commands: 'save log', 'clear', 'help'.", 'cyan')
            continue

        if not user_name:
            extracted = extract_name(user_input)
            if extracted:
                user_name = extracted
                bot_say(f"Nice to meet you, {user_name}! 😊", 'green')
                continue

        if is_farewell(user_input):
            reply, _ = match_intent(user_input)
            bot_say(reply, 'cyan')
            break

        reply, _ = match_intent(user_input)
        bot_say(reply, 'cyan')

    if conversation_log:
        choice = input("\n💾 Save conversation? (y/n): ").strip().lower()
        if choice == 'y':
            save_conversation()
    print_colored("👋 Thanks for chatting! Have a great day.", 'green', bold=True)

# ----------------------------
# 4.5.8. Run the chatbot
# ----------------------------
if __name__ == "__main__":
    chat_polished()
````

````python
# ==================================================
# 4.6. Saving & Documenting the Project (Basic Chatbot)
# ==================================================

import os
import zipfile
from datetime import datetime
from google.colab import files

# Create results directory
if not os.path.exists('chatbot_results'):
    os.makedirs('chatbot_results')
print("📁 Created directory: chatbot_results/")

# ----------------------------
# 4.6.1 Save the complete chatbot as a standalone Python script
# ----------------------------
chatbot_script = '''# Basic Chatbot – CodeAlpha Task 4
# A rule‑based conversational agent with coloured output, typing effect, and conversation logging.

import random
import re
import time
import sys
import datetime

# ---------- Coloured output ----------
def print_colored(text, color='white', bold=False, end='\\n'):
    colors = {
        'red': '\\033[91m', 'green': '\\033[92m', 'yellow': '\\033[93m',
        'blue': '\\033[94m', 'purple': '\\033[95m', 'cyan': '\\033[96m',
        'white': '\\033[97m', 'reset': '\\033[0m'
    }
    bold_code = '\\033[1m' if bold else ''
    reset_code = colors['reset']
    color_code = colors.get(color, colors['white'])
    print(f"{bold_code}{color_code}{text}{reset_code}", end=end)

# ---------- Typing effect ----------
def type_like_bot(text, delay=0.03):
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()

# ---------- Conversation logging ----------
conversation_log = []

def log_message(sender, message):
    timestamp = datetime.datetime.now().strftime("%H:%M:%S")
    conversation_log.append(f"[{timestamp}] {sender}: {message}")

def save_conversation(filename=None):
    if not conversation_log:
        print_colored("No conversation to save.", 'yellow')
        return
    if filename is None:
        filename = f"chat_log_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
    with open(filename, 'w', encoding='utf-8') as f:
        f.write("\\n".join(conversation_log))
    print_colored(f"💾 Conversation saved to {filename}", 'green')
    try:
        from google.colab import files
        files.download(filename)
    except:
        pass

# ---------- Intent definitions ----------
intents = [
    {
        "keywords": ["hello", "hi", "hey", "greetings", "good morning", "good afternoon", "good evening"],
        "responses": [
            "Hi there! 😊 How can I help you today?",
            "Hello! 👋 What's on your mind?",
            "Hey! 😄 Ready to chat?"
        ],
        "priority": 1
    },
    {
        "keywords": ["how are you", "how are u", "how do you do", "how's it going", "how's everything"],
        "responses": [
            "I'm just a bunch of code, but I'm doing great! Thanks for asking. 🤖",
            "I'm functioning perfectly! How about you? 😊",
            "All systems operational! How can I assist you?"
        ],
        "priority": 2
    },
    {
        "keywords": ["your name", "who are you", "what are you", "introduce yourself"],
        "responses": [
            "I'm ChatBot, your virtual assistant! 🤖",
            "I'm a simple rule‑based chatbot created for CodeAlpha. Nice to meet you!"
        ],
        "priority": 1
    },
    {
        "keywords": ["are you real", "real", "alive"],
        "responses": [
            "As real as Python code can be! 😉",
            "I exist in the digital realm. Pretty real to me!"
        ],
        "priority": 1
    },
    {
        "keywords": ["what can you do", "capabilities", "features", "help me"],
        "responses": [
            "I can chat about simple topics. Try saying 'hello', 'how are you', or 'bye'!"
        ],
        "priority": 1
    },
    {
        "keywords": ["help", "commands", "what to say", "options"],
        "responses": [
            "I understand greetings, 'how are you', 'your name', 'bye', and a few more. Just type naturally!"
        ],
        "priority": 2
    },
    {
        "keywords": ["thanks", "thank you", "appreciate", "good bot"],
        "responses": [
            "You're very welcome! 😊",
            "My pleasure! 🤗",
            "Anytime! 😊"
        ],
        "priority": 1
    },
    {
        "keywords": ["bye", "goodbye", "exit", "quit", "see you", "farewell"],
        "responses": [
            "Goodbye! Have a wonderful day! 👋",
            "Exiting chat. Take care! 👋",
            "Quitting. See you next time! 👋"
        ],
        "priority": 3
    }
]

DEFAULT_RESPONSE = "I'm not sure how to respond to that. 🤔 Try saying 'hello', 'how are you', or 'help'."

def match_intent(user_input):
    user_input_lower = user_input.lower()
    sorted_intents = sorted(intents, key=lambda x: x.get('priority', 0), reverse=True)
    for intent in sorted_intents:
        for keyword in intent["keywords"]:
            if keyword in user_input_lower:
                reply = random.choice(intent["responses"])
                return reply, intent
    return DEFAULT_RESPONSE, None

def is_farewell(user_input):
    reply, intent = match_intent(user_input)
    if intent and any(kw in user_input.lower() for kw in ["bye", "goodbye", "exit", "quit", "see you", "farewell"]):
        return True
    return False

# ---------- Name extraction ----------
user_name = None

def extract_name(user_input):
    patterns = [
        r"my name is (\\w+)",
        r"i am (\\w+)",
        r"call me (\\w+)",
        r"i'm (\\w+)"
    ]
    for pattern in patterns:
        match = re.search(pattern, user_input.lower())
        if match:
            return match.group(1).capitalize()
    return None

def get_personalised_greeting():
    if user_name:
        return f"Welcome back, {user_name}! 😊 How can I help you today?"
    else:
        return "Hi there! 😊 What's your name? (You can say 'My name is ...')"

# ---------- Bot speaking ----------
def bot_say(message, color='cyan', delay=0.03):
    print_colored("Bot: ", color, bold=True, end='')
    type_like_bot(message, delay)
    log_message("Bot", message)

# ---------- Main chat loop ----------
def chat_polished():
    global user_name
    print_colored("\\n" + "="*50, 'cyan', bold=True)
    print_colored("🤖 WELCOME TO THE ADVANCED CHATBOT! 🤖", 'green', bold=True)
    print_colored("="*50, 'cyan', bold=True)
    print("\\n💡 Tips:")
    print("   • Type 'save log' to save the conversation.")
    print("   • Type 'clear' to clear the screen (simulated).")
    print("   • Type 'help' for commands.")
    print("   • Type 'bye' to exit.\\n")

    greeting = get_personalised_greeting()
    bot_say(greeting)

    while True:
        print_colored("You: ", 'green', end='')
        user_input = input().strip()

        if not user_input:
            bot_say("Please say something! 😊", 'yellow')
            continue

        log_message("You", user_input)

        cmd = user_input.lower()
        if cmd == 'save log':
            save_conversation()
            continue
        elif cmd == 'clear':
            print("\\n" * 2)
            continue
        elif cmd == 'help':
            bot_say("I understand greetings, 'how are you', 'your name', 'bye'. Also commands: 'save log', 'clear', 'help'.", 'cyan')
            continue

        if not user_name:
            extracted = extract_name(user_input)
            if extracted:
                user_name = extracted
                bot_say(f"Nice to meet you, {user_name}! 😊", 'green')
                continue

        if is_farewell(user_input):
            reply, _ = match_intent(user_input)
            bot_say(reply, 'cyan')
            break

        reply, _ = match_intent(user_input)
        bot_say(reply, 'cyan')

    if conversation_log:
        choice = input("\\n💾 Save conversation? (y/n): ").strip().lower()
        if choice == 'y':
            save_conversation()
    print_colored("👋 Thanks for chatting! Have a great day.", 'green', bold=True)

if __name__ == "__main__":
    chat_polished()
'''

with open('chatbot_results/basic_chatbot.py', 'w') as f:
    f.write(chatbot_script)
print("💾 Saved standalone Python script: basic_chatbot.py")

# ----------------------------
# 4.6.2 Generate README.md for GitHub
# ----------------------------
readme_content = (
    "# Basic Chatbot – CodeAlpha Task 4\n\n"
    "## 📌 Project Overview\n"
    "A simple rule‑based conversational agent built in Python. The chatbot responds to common phrases like greetings, "
    "‘how are you’, ‘your name’, and ‘bye’. It features coloured output, a typing effect, conversation logging, "
    "and name memory for a personalised experience.\n\n"
    "## 🎮 Features\n"
    "- Responds to greetings, well‑being questions, farewells, and thanks.\n"
    "- Randomised replies for variety.\n"
    "- Coloured console output (cyan for bot, green for user).\n"
    "- Typing animation (simulates human typing).\n"
    "- Conversation logging with timestamps (save to `.txt` file).\n"
    "- User name extraction and memory (e.g., ‘My name is Alex’).\n"
    "- Command system: `save log`, `clear`, `help`.\n"
    "- Loop continues until user says ‘bye’, ‘exit’, or ‘quit’.\n\n"
    "## 🛠️ Tech Stack\n"
    "- Python 3.x (standard library only – no external dependencies)\n"
    "- Google Colab / any terminal with Python\n\n"
    "## 🚀 How to Use\n"
    "1. Run the script: `python basic_chatbot.py`\n"
    "2. Type your messages and press Enter.\n"
    "3. Say ‘hello’ to start, ‘how are you’ to ask, ‘bye’ to exit.\n"
    "4. Use `save log` to save the conversation, `clear` to clear the screen.\n\n"
    "## 📁 Repository Structure\n"
    "```\n"
    "CodeAlpha_Basic_Chatbot/\n"
    "├── basic_chatbot.py          # Main script\n"
    "├── Basic_Chatbot.ipynb       # Colab notebook (if used)\n"
    "├── README.md                 # This file\n"
    "├── requirements.txt          # (empty, no dependencies)\n"
    "└── chatbot_results/          # Output folder (optional)\n"
    "```\n\n"
    "## 🖼️ Sample Conversation\n"
    "```\n"
    "Bot: Hi there! 😊 What's your name? (You can say 'My name is ...')\n"
    "You: My name is Alex\n"
    "Bot: Nice to meet you, Alex! 😊\n"
    "You: How are you?\n"
    "Bot: I'm just a bunch of code, but I'm doing great! Thanks for asking. 🤖\n"
    "You: bye\n"
    "Bot: Goodbye! Have a wonderful day! 👋\n"
    "```\n\n"
    "## 📽️ Video Explanation\n"
    "[Link to LinkedIn video] – Replace with your actual URL.\n\n"
    "## 👩‍💻 Author\n"
    "Your Name – [LinkedIn Profile](https://linkedin.com/in/yourprofile)\n\n"
    "## 🙏 Acknowledgements\n"
    "CodeAlpha for the internship opportunity.\n\n"
    "---\n"
    "⭐ Star this repo if you found it useful!\n"
)

with open('README.md', 'w') as f:
    f.write(readme_content)
print("📝 README.md created for GitHub")

# ----------------------------
# 4.6.3 Create requirements.txt (no external packages)
# ----------------------------
with open('requirements.txt', 'w') as f:
    f.write("# No external dependencies required for Basic Chatbot\n# Python standard library only\n")
print("📦 requirements.txt created")

# ----------------------------
# 4.6.4 Create a sample conversation output (optional)
# ----------------------------
sample_convo = """Sample conversation log (example):
[14:30:15] Bot: Hi there! 😊 What's your name? (You can say 'My name is ...')
[14:30:22] You: My name is Alex
[14:30:22] Bot: Nice to meet you, Alex! 😊
[14:30:28] You: How are you?
[14:30:28] Bot: I'm just a bunch of code, but I'm doing great! Thanks for asking. 🤖
[14:30:35] You: bye
[14:30:35] Bot: Goodbye! Have a wonderful day! 👋
"""
with open('chatbot_results/sample_conversation.txt', 'w') as f:
    f.write(sample_convo)
print("📄 Sample conversation saved for reference")

# ----------------------------
# 4.6.5 Zip and download (fixed name clash)
# ----------------------------
from google.colab import files as colab_files   # use alias to avoid conflict

zip_path = 'basic_chatbot_results.zip'
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for root, dirs, files in os.walk('chatbot_results'):
        for file in files:
            zipf.write(os.path.join(root, file), arcname=os.path.join('chatbot_results', file))
    zipf.write('README.md')
    zipf.write('requirements.txt')
    if os.path.exists('Basic_Chatbot.ipynb'):
        zipf.write('Basic_Chatbot.ipynb')

print(f"🗜️ Zipped all results to {zip_path}")
colab_files.download(zip_path)   # now using alias
print("✅ Download started. Your results zip file is being saved locally.")
# ----------------------------
# 4.6.6 Final Completion Message
# ----------------------------
print("\n" + "="*60)
print("🎉 TASK 4 (BASIC CHATBOT) COMPLETED SUCCESSFULLY! 🎉")
print("="*60)
print("All results saved in 'chatbot_results/' folder and zipped for download.")
print("\n📌 Next Steps for Submission:")
print("1. Upload the notebook/script to GitHub repository: CodeAlpha_Basic_Chatbot")
print("2. Add README.md and requirements.txt to the repo")
print("3. Add the 'chatbot_results' folder contents (or at least the sample conversation)")
print("4. Record a 1-1.5 minute video demonstrating the chatbot (greeting, name memory, conversation, save log)")
print("5. Post video on LinkedIn with @CodeAlpha tag and GitHub link")
print("6. Submit the form shared in the WhatsApp group")
print("\n🌟 Great work on completing Task 4!")
````
